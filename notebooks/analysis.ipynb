```python
import pandas as pd

url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)

df.head()
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 1000

df = pd.DataFrame({
    "age": np.random.randint(20, 80, n),
    "smoker": np.random.binomial(1, 0.3, n),
})

# create hypertension with dependency on smoking + age
df["hypertension"] = (
    0.03 * df["age"] + 
    0.8 * df["smoker"] + 
    np.random.normal(0, 1, n)
)

df["hypertension"] = (df["hypertension"] > 2.5).astype(int)

df.head()
```

```python
import statsmodels.api as sm

# define variables
X = df[["smoker", "age"]]
X = sm.add_constant(X)

y = df["hypertension"]

# fit model
model = sm.Logit(y, X).fit()

print(model.summary())
```

```python
print("Interpretation:")

print("- Smoking is positively associated with hypertension.")
print("- Age is also positively associated with hypertension.")
print("- Both effects are statistically significant (p < 0.001).")
```

## Interpretation of Results

We performed a logistic regression analysis to evaluate the association between smoking and hypertension while adjusting for age.

The results indicate that:

- Smoking is associated with a higher likelihood of hypertension.
- Increasing age is also associated with higher risk of hypertension.
- Both variables are statistically significant (p < 0.001).

These findings suggest that smoking may be an important risk factor for hypertension, although causal interpretation is limited due to the observational nature of the data.
